```python
import numpy as np
import math as m
from sklearn import metrics
import sklearn
from sklearn.cross_validation import KFold
import urllib
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import cvxopt as cv
import pandas as pd
from sklearn import svm
import cvxopt.solvers
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import train_test_split
%matplotlib inline

def f(x, w, b, c=0):
            return (-w[0] * x - b + c) / w[1]
        
def cal_w0(X_train,y_train, L,w):
    no_of_sv = len(L[L!=0])
    print no_of_sv
    w0 =0
    for i in range(len(X_train)):
        w0 = w0 + (L[i]* (y_train[i]- (np.dot(w ,X_train[i,]))))
    return w0
#     return(float(1/no_of_sv)*w0)

def cal_wtx(X_train,y_train, L,K):
   
    wtx = 0
    for i in range(len(X_train)):
        wtx = wtx - (L[i]*y_train[i]* K[i])
    return wtx

def poly(x, y, p=2):
    return (1 + np.dot(x, y)) ** p


def gauss(x, y, sigma=5.0):
    return np.exp(-np.linalg.norm(x-y)**2 / (2 * (sigma ** 2)))
    
def main():
   
    iris = load_iris()
    X1 = iris.data
    y1 = iris.target
    X = iris.data[50:150:,0:2] 
    y = iris.target[50:150]
    for i in range(len(y)):
        if y[i]==2:
            y[i]=-1
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
#     print X_train[:,0]
    x_train1= pd.DataFrame(columns=(range(2)))
    x_train2= pd.DataFrame(columns=(range(2)))
    for i in range(len(X_train)):
        if y_train[i]==-1:
            x_train1.loc[i]=X_train[i]
        elif y_train[i]==1:
            x_train2.loc[i]=X_train[i]
    x_train1= np.matrix(x_train1)
    x_train2= np.matrix(x_train2)
   
    n_samples, n_features = X_train.shape
    K = np.zeros((n_samples, n_samples))
    Gaussian=True
    if Gaussian:
        for i in range(n_samples):
            for j in range(n_samples):
                K[i,j] = gauss(X[i], X[j])
    else:
        for i in range(n_samples):
            for j in range(n_samples):
                K[i,j] = poly(X[i], X[j])

    m= np.dot(y_train, y_train.T)
    b=np.dot(X_train,X_train.T)
    P= cv.matrix(np.dot(m,b))
#     q=np.full(n_samples, -1)
    q = cv.matrix(np.ones(n_samples) * -1)
    h = cv.matrix(np.zeros(n_samples))
    G = cv.matrix(np.diag(np.ones(n_samples) * -1))
    A= cv.matrix(y_train.astype(float),(1,n_samples))
    b=cv.matrix(0.0)
    solution = cvxopt.solvers.qp(P, q, G, h, A, b)
    L = np.ravel(solution['x'])
#     for i in range(len(L)):
#         if L[i] > 1e-8:
#             L[i]=0.00
#     print L
    sv = L > 1e-8
    ind = np.arange(len(L))[sv]
    L = L[sv]
#     print L
    sv_x = X_train[sv]
    sv_y = y_train[sv]
    print "%d support vectors out of %d points" % (len(L), n_samples)
    b = 0
#     print L.shape, sv_y.shape, K[ind[0],sv].shape
    for n in range(len(L)):
        b += sv_y[n]
#         print a.dtype,sv_y.astype(float).dtype, sv_y.dtype, K[ind[n].astype(int),sv.astype(int)].T.dtype
#         b = b- np.sum(np.dot(L, np.dot(sv_y, K[ind[n].astype(int),sv.astype(int)].T)))
        b -= np.sum(L * sv_y * K[ind[n],sv])
#         b=b-m
    b /= len(L)

#     # Weight vector
#     print L, sv_y, sv
    w = np.zeros(n_features)
    for n in range(len(L)):
        w += L[n] * sv_y[n] * sv_x[n]
#     print L[0] * sv_y[0] * sv[0]
    y_pred= np.sign(np.dot(X_test, w) + b)
    correct = np.sum(y_pred == y_test)
    print "%d out of %d predictions correct" % (correct, len(y_pred))
    d=pd.DataFrame(columns=('y_test','y_pred'))
    for i in range(len(y_test)):
        d.loc[i]=(y_test[i],y_pred[i])
#     print d
    clf = svm.SVC()
    clf.fit(X_train, y_train) 
    sv1=clf.support_vectors_
    fun_pred=clf.predict(X_test)
    Fun_Accuracy = metrics.accuracy_score(y_test, fun_pred)
    print "ACCURACY BY FUNCTION", Fun_Accuracy
    Accuracy = metrics.accuracy_score(y_test, y_pred)
    print "ACCURACY BY MY FUNCTION", Accuracy
#     print b
    plt.plot(x_train1[:,0], x_train1[:,1], "ro")
    plt.plot(x_train2[:,0], x_train2[:,1], "bo")
    plt.scatter(sv_x[:,0], sv_x[:,1], s=100, c="black")
    plt.savefig("kernel.png")
    plt.show()
# #     print w, b
#     plt.plot(x_train1[:,0], x_train1[:,1], "ro")
#     plt.plot(x_train2[:,0], x_train2[:,1], "bo")
#     plt.scatter(sv_x[:,0], sv_x[:,1], s=100, c="black")
#     a0 = 5; a1 = f(a0, w, b)
#     b0 = 5; b1 = f(b0, w, b)
#     plt.plot([a0,b0], [a1,b1], "k")
#     plt.axis("tight")
# #     plt.axis([4, 7.5, 1.5, 5])
#     plt.show()
#     w = np.zeros(n_features)
#     for i in range(len(L)):
#         w =w+  L[i] * y_train[i] * X_train[i,]
# #     print w
#     w0= cal_w0(X_train,y_train, L,w)
# #     print w0
#     wtx=cal_wtx(X_train,y_train, L,K)
# #     print wtx
#     y_pred=[]
#     for i in range(len(X_train)):
#         val = wtx[i] + w0
#         print val
#         if(val>0):
#             res= 1
#             y_pred.append(res)
#         else:
#             res=-1
#             y_pred.append(res)
#     print y_pred


    
if __name__ == "__main__":main()
```

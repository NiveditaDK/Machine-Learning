```python
from sklearn import datasets, linear_model
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
import sklearn
from sklearn.cross_validation import KFold
from sklearn.cross_validation import train_test_split
import numpy as np
from math import exp
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

# def log_reg(z,y):
#     theta = [0.5,-0.1,0.1,-0.5,0.3]
#     combine_theta=[]
#     for i in range(1000):
#         combine_theta=np.append(combine_theta,theta)
#         y_hat_set=[]
        
#         for i in range(len(y)):
#             y_hat = (1/(1+exp(-(np.dot(np.array([z[i,]]),theta)))))
#             y_hat_set = np.append(y_hat_set, y_hat)
#             pred_op = y_hat_set
#         alpha = 0.1
#         temp = 0
#         tl = np.zeros(5)
#         for i in range(len(y)):
#             temp = theta - alpha * 1/len(z) * (y_hat_set[i] - y[i]) * z[i,]
#             tl = np.append(tl,temp.T)
#         theta = tl
# #         tl1=pd.DataFrame(columns=('1', '2','3','4','5'))
# #         tl1.loc[1] = (min(theta[1]), min(theta[2]),min(theta[3]),min(theta[4]),min(theta[5]))
# #         theta = tl1.values
#     print theta


def h_th(j,theta ,z):
    a = exp(np.dot(theta[j].T,z))
    b = sum([exp(np.dot(theta[j].T,z)) for i in range(len(z))])
    h_theta = a/b
    return h_theta

def predict(z_test,theta,y_test,classes):
    df=pd.DataFrame(columns=('y_pred','y_test'))
    pred=[]
    index = 0
    for i in z_test:
        temp_theta =[]
        for j, value in enumerate(classes):
            parameter = h_th(j, theta,np.array([i]).T)
            temp_theta.append(parameter)
        max_value = max(temp_theta)
        for i, value in enumerate(temp_theta):
            if (value == max_value):
                index = i
                break
        pred.append(index)
#     df[y_pred]=pred
#     df[y_test]=y_test
#     print np.array(pred), y_test
    return np.array(pred), y_test
#     print df
#     return df

def main():
    iris = load_iris()
    X = iris.data 
    Y = iris.target
    print('LOGISTIC REGRESSION FOR K CLASS')
    fold=0
    kf=sklearn.cross_validation.KFold(n=len(X), n_folds=10, shuffle=False,random_state=None)
    for train_index, test_index in kf:
        fold+=1
        X_train,X_test=X[train_index], X[test_index]
        y_train,y_test=Y[train_index], Y[test_index]
        z_train = PolynomialFeatures(1).fit_transform(X_train)
        z_test= PolynomialFeatures(1).fit_transform(X_test)
        classes=np.unique(Y)
        theta =[np.array([np.ones(z_train.shape[1])]).T/100 for i in classes]
        for i, value in enumerate(classes): #for each class
            for j in range(len(z_train)): #for each sample
                for k in range(z_train.shape[1]): #for each feature
                    if (y_train[j] == value):
                        ind=y_train[j]*1.0
                    else:
                        ind=0
                    theta[i][k]=theta[i][k]-(0.001*(h_th(i,theta,np.array([z_train[j]]).T)-ind)*np.array(z_train[j][k]))  
#         print theta
        print "FOR FOLD:", fold
        y_pred, y_act= predict(z_test,theta,y_test,classes)
        print 'Predicted Y : ' ,y_pred
        print 'Actual Y : ' ,y_act
        Accuracy = metrics.accuracy_score(y_act, y_pred)
        print "ACCURACY BY MY MODEL" ,Accuracy
        fun =   LogisticRegression()
        fun.fit(z_train,y_train)
        fun_pred = fun.predict(z_test)
        Fun_Accuracy = metrics.accuracy_score(y_act, fun_pred)
        print "ACCURACY BY FUNCTION", Fun_Accuracy
#         print "y_pred" , "y_test"
#         print y_pred, y_test
#        
#         return theta

#         calculate_parameters(z_train,y_train,theta)
#         print theta
#         y_1=range(0,len(y_train))
#         y_2=range(0,len(y_train))
#         y_3=range(0,len(y_train))
#         for i in y_1:
#             if (y_train[i]==0):
#                 y_1[i]=1
#             else:
#                 y_1[i]=0
#         for i in y_2:
#             if (y_train[i]==1):
#                 y_2[i]=1
#             else:
#                 y_2[i]=0
#         for i in y_3:
#             if (y_train[i]==2):
#                 y_3[i]=1
#             else:
#                 y_3[i]=0
#         log_reg(z_train,y_1)
#         print y_1, y_2, y_3
        
    
if __name__ == "__main__":main()
```

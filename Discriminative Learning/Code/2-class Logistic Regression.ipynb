```python
from sklearn import datasets, linear_model
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
import sklearn
from sklearn.cross_validation import KFold
from sklearn.cross_validation import train_test_split
import numpy as np
from math import exp
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

def h_th(a,b):
#     h_theta=1.0/(1.0 + exp(-1.0 *(float(np.dot(theta.T,p)))))
    h_theta= 1.0/(exp(-1.0 * float(np.dot(b.T,a)))+1.0)
    return h_theta
    
def calculate_parameters(z,Y_train,theta):
   
    d=pd.DataFrame(columns=('1', '2','3','4','5'))
    cost=pd.DataFrame(columns=('1','2','3','4','5'))
    y_train=range(0,len(Y_train))
    for i in y_train:
        if (Y_train[i]==1):
            y_train[i]=1
        else:
            y_train[i]=0
    for i, value in enumerate(z):
#         print theta, np.array([value])
#         inp= float(np.dot(theta.T,np.array([value])))
#         h_theta= h_th(inp)
        h_theta=h_th(np.array([value]).T,theta)
#         print h_theta, y_train
        l_theta=(h_theta - y_train[i]) * np.array(value)
        d.loc[i]=l_theta
        cost.loc[1]=d.sum(axis=0)
#     print cost
    li=cost.values
    return li
#     for i in range(iterations):
#         theta=theta - (learning_rate * li)
#     return theta

    
def predict(z_test,theta,y_test):  
    df=pd.DataFrame(columns=('y_pred','y_test'))
    y_act=range(0,len(y_test))
    for i in y_act:
        if (y_test[i]==1):
            y_act[i]=1
        else:
            y_act[i]=0
    for i, value in enumerate(z_test):
#         print  theta, np.array([z_test[i,]])
        h_theta=h_th(np.array([value]).T,theta)
#         print h_theta
        if(h_theta>0.5):
            c=1
        elif(h_theta<0.5):
            c=0
        df.loc[i]=(c,y_act[i])
    return df
    
        
def main():
    iris = load_iris()
    X = iris.data 
    Y = iris.target
    enc = LabelEncoder()
    label_encoder = enc.fit(Y)
    y = label_encoder.transform(Y) + 1
    x=X[0:100]
    y=y[0:100]
    fold=0
    print('LOGISTIC REGRESSION FOR 2 CLASS')
#     X_train, X_test, y_train, y_test = train_test_split(x, y_new, test_size=0.30,random_state=42)
    kf=sklearn.cross_validation.KFold(n=len(x), n_folds=10, shuffle=True,random_state=None)
    for train_index, test_index in kf:
        fold+=1
        X_train,X_test=x[train_index], x[test_index]
        y_train,y_test=y[train_index], y[test_index]
        classes=np.unique(y)
        z_train = PolynomialFeatures(1).fit_transform(X_train)
        z_test= PolynomialFeatures(1).fit_transform(X_test)
        theta = (np.array([np.ones(z_train.shape[1])]).T)/1000
#         print theta, z_train
        for i in range(500):
#             li=calculate_parameters(z_train,y_train,theta)
            theta=theta - (0.0001 * calculate_parameters(z_train,y_train,theta).T)
        df=predict(z_test,theta,y_test)
#     print theta
#         print theta
#         df=predict(z_test,theta,y_test)
        print('FOR FOLD', fold)
        print df
        Accuracy = metrics.accuracy_score(df.y_test, df.y_pred)
        print "ACCURACY BY MY MODEL" ,Accuracy
        fun =   LogisticRegression()
        fun.fit(z_train,y_train)
        fun_pred = fun.predict(z_test)
        Fun_Accuracy = metrics.accuracy_score(y_test, fun_pred)
        print "ACCURACY BY FUNCTION", Fun_Accuracy
#     print classes
#     print x
    
if __name__ == "__main__":main()
```

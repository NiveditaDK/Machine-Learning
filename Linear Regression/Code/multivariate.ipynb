```python
import numpy as np
import urllib
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as m_s_e
from sklearn import linear_model
from mpl_toolkits.mplot3d import Axes3D
import sklearn
from sklearn.cross_validation import KFold
from sklearn.preprocessing import PolynomialFeatures as pf
%matplotlib inline

# Ready made function for Linear Regression
def readymade(x,y,x_test,y_test):
    mse1=0
    regr = linear_model.LinearRegression()
    regr.fit(x,y)
    prediction= regr.predict(x_test)
    mse1=mse1+m_s_e(y_test,prediction)
    return mse1

# Function for 10 fold cross validation
def k_fold(X, Y,Z, k_fold,deg):
    mse=0
    th=0
    err=0
    errForReadymade=0
    mse1=0
    mte=0
    err1=0
    mteste=0
    err2=0
    kf=sklearn.cross_validation.KFold(n=len(X), n_folds=10, shuffle=False,random_state=None)
    for train_index, test_index in kf:
        X_train,X_test=X[train_index], X[test_index]
        Z_train,Z_test=Z[train_index], Z[test_index]
        Y_train,Y_test=Y[train_index], Y[test_index]
        mse,theta=fit_model(deg,Z_train,Y_train,Z_test,Y_test)
        mte=fit_model_train(Z_train,Y_train)
        mteste=fit_model_test(Z_test,Y_test)
        mse1=readymade(X_train,Y_train,X_test,Y_test)
        err=err+mse
        err1=err1+mte
        err2=err2+mteste
        errForReadymade=errForReadymade+mse1
        th=th+theta
    err2=err2/k_fold
    err1=err1/k_fold
    err=err/k_fold
    errForReadymade=errForReadymade/k_fold
    th=th/k_fold
    print 'Theta:',th    
    print 'MSE with defined algorithm:',err
    print 'MSE With Readymade Function: ',errForReadymade
    print 'MTE with defined algorithm: ',err1
    print 'MTestE with defined algorithm: ',err2


# Function to define model on training and applying on test   
def fit_model(deg,z,y,z_test,y_test):
    mse=0
    theta=np.dot(np.linalg.pinv(z),y)
#     print theta
    y_hat= np.dot(theta.transpose(),z_test.transpose())
#     print y_hat
    mse=mse+m_s_e(y_test,y_hat)
    return mse,theta
    
# Fit model on training data set
def fit_model_train(z,y):
    mte=0
    theta=np.dot(np.linalg.pinv(z),y)
#     print theta
    y_hat= np.dot(theta.transpose(),z.transpose())
#     print y_hat
    mte=mte+m_s_e(y,y_hat)
    return mte

# Fit model on training data set
def fit_model_test(z,y):
    mteste=0
    theta=np.dot(np.linalg.pinv(z),y)
#     print theta
    y_hat= np.dot(theta.transpose(),z.transpose())
#     print y_hat
    mteste=mteste+m_s_e(y,y_hat)
    return mteste

# Function for iterative solution
def fit_modelForgradient(X,Y,Z, k_fold,deg):
    thetawithgradient=0
    errorwithgradient=0
    kf=sklearn.cross_validation.KFold(n=len(X), n_folds=10, shuffle=False,random_state=None)
    for train_index, test_index in kf:
        X_train,X_test=X[train_index], X[test_index]
        Z_train,Z_test=Z[train_index], Z[test_index]
        Y_train,Y_test=Y[train_index], Y[test_index]
        m = np.shape(Z_train)
        alpha = 0.001 
        theta1,mse2,y_hat = gradient_descent(alpha, Z_train, Y_train,Z_test,Y_test, 2000,deg)
        thetawithgradient=thetawithgradient+theta1
        errorwithgradient=errorwithgradient+mse2
    errorwithgradient=errorwithgradient/k_fold
    thetawithgradient=thetawithgradient/k_fold
    print 'Theta with gradient:',thetawithgradient
    print 'MSE with gradient Descent:' ,errorwithgradient
    return Z_test,Y_test,X_test,y_hat

def gradient_descent(alpha, z, y,z_test,y_test, numIterations,deg):
    mse=0
    m = z.shape[0]
    d=z.shape[1]
    theta = np.random.rand(d)
    z_transpose = z.transpose()
    for iter in range(0, numIterations):
        hypothesis = np.dot(z, theta)
        loss = hypothesis - y
        J = np.sum(loss ** 2) / (2 * m)      
        gradient = np.dot(z_transpose, loss) / m    
#         print theta.shape, gradient.shape
        theta = theta - (alpha * gradient) 
#     if(deg==1)
#     for i in range(z_test.shape[1]):
#             y_hat = theta[0] + theta[1]*z_test +theta[2]*(np.square(z_test))        
    y_hat=np.dot(theta.T,z_test.T)
    mse=mse+m_s_e(y_test,y_hat)
    return theta,mse,y_hat

# function to plot data after applying model
def plot_later(z,y,x,pred,name,d): 
    plt.plot(x,y,'bo',label='Actual')
    plt.plot(x,pred,'ro',label='Predicted')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('For One of the folds')
    plt.legend()
    plt.grid(True)
    if(name=='mvar-set1.dat'):
        plt.savefig("plotLater"+name+str(d)+".png")
    if(name=='mvar-set2.dat'):
        plt.savefig("plotLater"+name+str(d)+".png")
    if(name=='mvar-set3.dat'):
        plt.savefig("plotLater"+name+str(d)+".png")
    if(name=='mvar-set4.dat'):
        plt.savefig("plotLater"+name+str(d)+".png")
    plt.show() 

def main():
    Deg = input("Enter Degree: ")
    url1 = "http://cs.iit.edu/~agam/cs584/data/regression/mvar-set1.dat"
    url2="http://cs.iit.edu/~agam/cs584/data/regression/mvar-set2.dat"
    url3="http://cs.iit.edu/~agam/cs584/data/regression/mvar-set3.dat"
    url4="http://cs.iit.edu/~agam/cs584/data/regression/mvar-set4.dat"
    urlList=[url1,url2]
    urlList1=[url3,url4]
    for data in urlList:  
        string=data.split("/")
        name=string[-1]
        print "For "+name
        data1 = urllib.urlopen(data)
        dataset = np.loadtxt(data1)
        X = dataset[:,0:2]
        Y = dataset[:,2]  
        poly=pf(Deg)
        Z=poly.fit_transform(X)
        np.set_printoptions(formatter={'float': '{: 0.5f}'.format})
        k_fold(X,Y,Z,10,Deg)
        z,y,x,y_hat=fit_modelForgradient(X,Y,Z,10,Deg)
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(X[:,0], X[:,1], Y, c='b', marker='o')
        ax.set_xlabel('X1 Label')
        ax.set_ylabel('X2 Label')
        ax.set_zlabel('Y Label')
        plt.savefig("plotInit"+name+".png")
        plt.show()
        plot_later(z,x,y,y_hat,name,Deg)
    for data in urlList1:  
        string=data.split("/")
        name=string[-1]
        print "For "+name
        data1 = urllib.urlopen(data)
        dataset = np.loadtxt(data1)
        X = dataset[:,0:5]
        Y = dataset[:,5]  
        plt.plot(X,Y,'bo')
        plt.savefig("plotInit"+name+".png")
        plt.show()
        poly=pf(Deg)
        Z=poly.fit_transform(X)
        np.set_printoptions(formatter={'float': '{: 0.5f}'.format})
        k_fold(X,Y,Z,10,Deg)
        z,y,x,y_hat=fit_modelForgradient(X,Y,Z,10,Deg)
        plot_later(z,x,y,y_hat,name,Deg)
#         fig = plt.figure()
#         ax = fig.add_subplot(111, projection='3d')
#         ax.scatter(X[:,0], X[:,1], Y, c='b', marker='o')
#         ax.set_xlabel('X1 Label')
#         ax.set_ylabel('X2 Label')
#         ax.set_zlabel('Y Label')
#         plt.show()
    
if __name__ == "__main__":main()    
```
